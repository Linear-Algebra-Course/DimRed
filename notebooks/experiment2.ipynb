```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import sys
sys.path.append('../dimred')
sys.path.append('../utils/')
import util
```

```python
hello, colorize = util.make_hello(1000)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(hello[:, 0], hello[:, 1], **colorize)
```

```python
hello_curve =util.make_hello_s_curve(hello)
```

```python
from mpl_toolkits import mplot3d
plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')
ax.scatter3D(hello_curve[:, 0], hello_curve[:, 1], hello_curve[:, 2],
             **colorize);
```

# PCA

```python
from pca import PCA
```

```python
pca = PCA(2)
```

```python
pca.fit(hello_curve)
```

```python
X_new = pca.transform(hello_curve)
```

```python
plt.figure(figsize=(8,4))
ax = plt.axes()
ax.scatter(X_new[:,0],X_new[:,1], **colorize);
```

# Random Projection

```python
from random_projection import RandomProjection
```

```python
random_proj = RandomProjection(n_component=2)
```

```python
random_proj.fit(hello_curve)
```

```python
X_new = random_proj.transform(hello_curve)
```

```python
plt.figure(figsize=(8,4))
ax = plt.axes()
ax.scatter(X_new[:,0],X_new[:,1], **colorize);
```

# Isomap

```python
from isomap import IsoMap
```

```python
isomap = IsoMap(n_component=2, n_neighbour=50)
```

```python
isomap.fit(hello_curve)
```

```python
X_new = isomap.transform(hello_curve)
```

```python
plt.figure(figsize=(8,4))
ax = plt.axes()
ax.scatter(X_new[:,0],X_new[:,1], **colorize);
```

# Laplacian method

```python
from laplacian import Laplacian
```

```python
laplacian = Laplacian(n_neighbours= 10, n_component=2)
```

```python
laplacian.fit(hello_curve)
```

```python
X_new = laplacian.transform()
```

```python
plt.figure(figsize=(8,4))
ax = plt.axes()
ax.scatter(X_new[:,0],X_new[:,1], **colorize);
```
